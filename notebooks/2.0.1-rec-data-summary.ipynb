## Getting to Know the Data for Analysis
The data used for analysis is the Christmas cbc circle data matched to at least 1 station. Additional data such as elivation and ecosystem is also included. 

A note on terminology:

A **Circle Location** is a unique latitude and logitude and is the physical center of a cbc circle where a count occured. 

A **CBC Count** is a Christmas Bird Count at a location in a given year. 

Thus several CBC counts can occure at the same circle location. 

Data that is stagment overtime (Ecosystem, eleivation) will be reported at the circle location level. 

Data that is unique to a year, like the weather, will be reported for the CBC Count. 
For the CBC counts data, a single record for the snow tempeture, snow depth is recorded. 
This data is compared to 1- n NOAA stations.

### How many stations share a ecosystem classification with a circle

```python
def compare_ecosys(x):
    circle_nlcd = x['Nlcd_circle']
    circle_usgs = x['Usgsid_sys_circle']
    
    station_nlcd = x['Nlcd_station']
    station_usgs = x['Usgsid_sys_station']
    
    if(~pd.isna(circle_nlcd)):
        nlcd_match = circle_nlcd == station_nlcd
    
    
    if(~pd.isna(circle_usgs)):
        usgs_match = circle_usgs == station_usgs
    
    
    return pd.Series({
        "nlcd_match": nlcd_match,
        "usgs_match": usgs_match
    })
    
    
```

```python
data[['nlcd_match', 'usgs_match']] = data.apply(compare_ecosys, axis = 1)
```

```python
print("The number of circles locations that have the same nlcd code as at least one reference station:")

data.loc[~pd.isna(data['Nlcd_circle']) & (data['nlcd_match'] == True)]['latlon_ser'].nunique()

```

```python
print("The number of circles that have the same usgs code as at least one reference station:")

data.loc[~pd.isna(data['Usgsid_sys_circle']) & (data['usgs_match'] == True)]['latlon_ser'].nunique()

```
